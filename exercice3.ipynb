```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Ames Housing.csv", sep=";", skiprows=1)
df.head()
```

2.1

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

Réponse 2.1

On calcule la matrice de corrélation pour les variables numériques.

La corrélation mesure la relation linéaire entre deux variables :
- proche de 1 → relation positive forte
- proche de -1 → relation négative forte
- proche de 0 → absence de relation

2.2

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.show()
```

 Réponse 2.2

La heatmap permet de visualiser les corrélations entre variables.

Elle permet d’identifier rapidement les relations importantes grâce aux couleurs.

```python
corr_target = corr_matrix["SalePrice"].sort_values(ascending=False)
corr_target
```

Réponse 2.3

On analyse la corrélation avec la variable cible `SalePrice`.

Cela permet d’identifier les variables les plus influentes sur le prix des maisons.

3.1

```python
corr_target[abs(corr_target) > 0.5]
```

Réponse 3.1

On sélectionne les variables ayant une corrélation forte avec `SalePrice`.

Cela permet de garder uniquement les variables les plus pertinentes pour la modélisation.

4.1

```python
selected_vars = corr_target[abs(corr_target) > 0.5].index
df_selected = df[selected_vars]

plt.figure(figsize=(8,6))
sns.heatmap(df_selected.corr(), annot=True, cmap="coolwarm")
plt.show()
```

 Réponse 4.1

On observe les corrélations entre les variables fortement liées à `SalePrice`.

Cela permet de détecter la multicolinéarité, c’est-à-dire lorsque des variables sont fortement corrélées entre elles.

La multicolinéarité peut poser problème dans les modèles statistiques.

4.2

```python
corr_selected = df_selected.corr()

corr_pairs = corr_selected.unstack()
corr_pairs = corr_pairs[abs(corr_pairs) > 0.8]
corr_pairs = corr_pairs[corr_pairs < 1]

corr_pairs
```

 Réponse 4.2

On identifie les paires de variables fortement corrélées entre elles (> 0.8).

Cela permet de repérer les variables redondantes qui apportent la même information.
