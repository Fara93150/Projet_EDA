```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Ames Housing.csv", sep=";", skiprows=1)
df.head()
```

2.1

```python
plt.figure(figsize=(8,5))
sns.histplot(df["SalePrice"], bins=30, kde=True)
plt.title("Distribution de SalePrice")
plt.show()
```

Réponse 2.1

On représente la distribution de `SalePrice` à l’aide d’un histogramme accompagné d’une courbe KDE.

Ce choix est adapté car il s’agit d’une variable quantitative.

Ce graphique permet d’observer :
- la forme de la distribution ;
- la concentration des valeurs ;
- une éventuelle asymétrie.

La courbe KDE permet de lisser la distribution pour mieux visualiser sa tendance.
