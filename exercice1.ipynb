```python
#### 2.2 Valeurs manquantes
##**Ce que vous devez faire :**
##Calculez le nombre de valeurs manquantes par colonne, ainsi que leur proportion (en pourcentage du total).
##Affichez uniquement les colonnes qui ont au moins une valeur manquante, triées par proportion décroissante.


# 1. On compte combien de "trous" (NaN) il y a par colonne
nb_manquants = df.isnull().sum()

# 2. On calcule la proportion (on divise par le nombre total de lignes)
# len(df) donne le nombre total de maisons dans ton fichier
prop_manquants = (nb_manquants / len(df)) * 100

# 3. On crée un petit tableau (DataFrame) pour y voir clair
df_manquants = pd.DataFrame({
    'Nombre': nb_manquants, 
    'Pourcentage %': prop_manquants
})

# 4. On garde SEULEMENT les colonnes qui ont au moins 1 trou
# On trie du plus gros pourcentage au plus petit (ascending=False)
df_final_manquants = df_manquants[df_manquants['Nombre'] > 0].sort_values(by='Pourcentage %', ascending=False)

# On affiche le résultat
display(df_final_manquants)
```

```python
###### 2.4 Sélection des variables d'intérêt
##  **Ce que vous devez faire :**#- Créez un DataFrame `df_subset` contenant uniquement les 5 variables ciblées (voir tableau ci-dessus). Gardez également une colonne identifiant les observations si elle existe.
# On définit la liste des noms des colonnes cibles
# (Attention à bien respecter les majuscules/minuscules)

# On définit la liste des noms des colonnes cibles
# (Attention à bien respecter les majuscules/minuscules)
# 1. On définit la liste EXACTE des noms (vérifie bien les espaces !)
mes_variables = ['Gr Liv Area', 'SalePrice', 'Lot Area', 'Year Built', 'Overall Qual']

# 2. On crée le mini-tableau
# (On utilise les noms qui sont RÉELLEMENT dans ton fichier CSV)
df_subset = df[mes_variables].copy()

# 3. On vérifie
df_subset.head()
```

```python
#SECTION 3
#3.1 à 3.4
df_subset.describe()
```

```python
# On prépare un dictionnaire pour stocker les résultats
stats_resultats = {}

for col in df_subset.columns:
    # --- Statistiques Classiques ---
    moyenne = df_subset[col].mean()
    std = df_subset[col].std()
    mini = df_subset[col].min()
    maxi = df_subset[col].max()
    
    # --- Statistiques Robustes ---
    mediane = df_subset[col].median()
    q1 = df_subset[col].quantile(0.25)
    q3 = df_subset[col].quantile(0.75)
    iqr = q3 - q1
    
    # MAD calculé à la main (logique du cours)
    mad = (df_subset[col] - mediane).abs().median()
    
    # --- Forme (Skewness & Kurtosis via Pandas) ---
    skewness = df_subset[col].skew()
    kurtosis = df_subset[col].kurtosis()
    
    stats_resultats[col] = {
        'Moyenne': moyenne, 'Écart-type': std, 'Min': mini, 'Max': maxi,
        'Médiane': mediane, 'Q1': q1, 'Q3': q3, 'IQR': iqr, 'MAD': mad, 
        'Skewness': skewness, 'Kurtosis': kurtosis
    }

# Création du tableau propre
df_recap = pd.DataFrame.from_dict(stats_resultats, orient='index')
display(df_recap)
```

1. Moyenne vs Médiane :
Regarde SalePrice : Normalement, la Moyenne est plus haute que la Médiane.
Pourquoi ? Parce qu'il y a quelques maisons extrêmement chères (les "châteaux") qui tirent la moyenne vers le haut. La médiane, elle, reste "au milieu" de la foule. C'est l'asymétrie positive.
2. Le Skewness (Asymétrie) :
Si le Skewness est très positif (ex: > 1), cela confirme qu'il y a une "longue queue" de valeurs élevées à droite du graphique. C'est typique du prix de l'immobilier ou de la surface du terrain (LotArea).
3. L'Écart-type vs le MAD :
L'écart-type est très sensible aux erreurs. Le MAD est beaucoup plus stable. Si l'écart-type est géant par rapport au MAD, c'est le signe qu'il y a beaucoup d'outliers (valeurs aberrantes) dans tes données.

Pour des variables comme SalePrice et LotArea, la moyenne est nettement supérieure à la médiane et le skewness est élevé (> 0). Cela indique une distribution étalée à droite avec la présence de valeurs extrêmes.
En revanche, pour YearBuilt, la moyenne et la médiane sont très proches, ce qui suggère une distribution plus symétrique.

```python
# On choisit la variable à afficher
col = 'SalePrice' 

# 1. On crée une figure avec 2 zones (1 ligne, 2 colonnes)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))

# --- GRAPHIQUE 1 : HISTOGRAMME ---
sns.histplot(df_subset[col], kde=True, ax=ax1, color='skyblue')
# On ajoute la moyenne (rouge) et la médiane (bleu)
ax1.axvline(df_subset[col].mean(), color='red', linestyle='-', label='Moyenne')
ax1.axvline(df_subset[col].median(), color='blue', linestyle='--', label='Médiane')
ax1.set_title(f'Distribution de {col}')
ax1.legend()

# --- GRAPHIQUE 2 : BOXPLOT ---
sns.boxplot(x=df_subset[col], ax=ax2, color='lightgreen')
ax2.set_title(f'Boxplot de {col}')

plt.show()
```
